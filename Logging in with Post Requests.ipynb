```python
# Accessing the profile page of Planespotters.net

# Multiple runs of this code might lead to login errors 
# as the website adds a layer of CAPTCHA when it detects multiple login attempts

#Importing the required packages
import requests
from bs4 import BeautifulSoup as bs
import time

#Accessing the login page

url_plane_spotter = "https://www.planespotters.net/user/login" #Storing the URL in a string

#Adding the header to mimic a browser
header = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36"}

## Starting a session
plane_spotter_session = requests.session()

#Storing the page from the session into a variable
plane_spotter_page1 = plane_spotter_session.get(url = url_plane_spotter, headers = header)


#Getting and storing the cookies from the session
cookies_page1 = plane_spotter_page1.cookies.get_dict()

#Parsing the HTML file of session 1
session1_parse = bs(plane_spotter_page1.text, 'lxml')

#We need to get the hidden "csrf" field that will be required in the login process
#This csrf field will then be used in a POST request with credentials
#New cookies received from the POST request will be addded to our cookies
#After this, a GET request will be sent with just the url, headers and cookies


#Getting the hidden field where the csrf is stored
csrf_input = session1_parse.find("input", id="csrf")

#Getting the csrf value
csrf = csrf_input.get("value")
#print(csrf)

#Adding sleep
time.sleep(5)

#Will now make a POST request with the credentials, cookies and the csrf
#This will log us into the website
plane_spotter_page2 = plane_spotter_session.post(url = url_plane_spotter, 
                                                headers = header,
                                               cookies = cookies_page1,
                                                data = {
                                                    "username" : "", #Username
                                                    "password" : "", #Password
                                                    "csrf" : csrf,
                                                    "rid" : ""
                                                }, timeout = 15)

#Getting the new cookies
cookies_page2 = plane_spotter_page2.cookies.get_dict()


#Parsing the logged in page
session2_parse = bs(plane_spotter_page2.text, 'lxml')

#Getting the new cookies
cookies_page2 = plane_spotter_page2.cookies.get_dict()

#Adding sleep
time.sleep(5)

#Creating a new variable for the new URL
url_plane_spotter_profile = "https://www.planespotters.net/member/profile"

#Updating the cookies to include all cookies
#cookies_page2.update(cookies_page1)

#Now, we will need to login to the profile using the saved cookies with a GET request
plane_spotter_page3 = plane_spotter_session.get(url = url_plane_spotter_profile, 
                                                headers = header,
                                                cookies = cookies_page2)

#Parsing the profile page
session3_parse = bs(plane_spotter_page3.text, 'lxml')

#Printing the Beautiful Soup document of my profile page
#print(session3_parse)

#Combining the cookies into a single dictionary to print
cookies_page2.update(cookies_page1)

#Printing the combined cookies
print(f"\n The combined cookies are as follows: \n {cookies_page2} \n")

#Checking if my username is present in the webpage
print(bool(session3_parse.findAll(text = "username")))
```
