```python
#Importing packages

import requests
from bs4 import BeautifulSoup as bs
import time
import re #RegEx package

#Adding the header to mimic a browser
header = {'user-agent': 'Chrome/99.0.4840.0'}

#### Part a
#Loading first page of B&N top 100 with 40 items

print("\n-------------Executing Part a-------------\n")

url_bn_a = "https://www.barnesandnoble.com/b/books/_/N-1fZ29Z8q8?Nrpp=40&page=1"

part_a = requests.get(url = url_bn_a, headers = header)

part_a_bs = bs(part_a.text, 'lxml')

#This variable stores the first page with 40 items

print("The first page has been stored in the variable part_a_bs.")

print("\n-------------Part a executed-------------\n")

#### Part b
#Getting the links of the 40 products in the first page

print("\n-------------Executing Part b-------------\n")

#Storing all the <a> tags of the product links in link_products list

products_list = part_a_bs.findAll("a", class_ = "format")

#Creating a list to store all the link
product_links = list(range(0,40))


for i in range(0,40):
    
    #Getting the link from the link tag
    product_links[i] = products_list[i].get("href")
    
    #Cleaning the link to remove the unnecessary part (The part starting with ";")
    product_links[i] = re.split(";",product_links[i])[0]
    
    #Concatenating with website to complete the link
    product_links[i] = f"https://www.barnesandnoble.com{product_links[i]}" 

print(f"Length of the product list is: {len(product_links)}\n") 
 
print(product_links)
      
print("\n-------------Part b executed-------------\n")

#### Part c
#Downloading all 40 product pages into computer
      
print("\n-------------Executing Part c-------------\n")

for i in range(1,41):
    
    #A temporary product page to store
    file_to_write = open(f"bn_top100_{i}.html",'w')
    
    #Storing the html of product page in a variable
    product_page = requests.get(url = product_links[i-1], headers = header)
    
    #Writing the html into the file
    file_to_write.write(product_page.text)
    
    #Adding a sleep time of 5 seconds before next request
    time.sleep(5)

print("All 40 product pages have been saved!")
      
print("\n-------------Part c executed-------------\n")
    
#### Part d
#Printing the first 100 characters of the "Overview" section of each product
      
print("\n-------------Executing Part d-------------\n")

for i in range(1,41):
    
    #Importing the file from the html
    product_page = open(f"bn_top100_{i}.html",'r')
    
    #Converting to a Beautiful Soup object
    product_page_bs = bs(product_page, 'lxml')
    
    #Getting the content from the class "overview-cntnt" which contains the "Overview" section
    text = product_page_bs.find("div", class_ = "overview-cntnt", itemprop = "description").text
    
    #Printing the first 100 characters from the "Overview" section of the page
    print(f"{text[0:100]}")
      
print("\n-------------Part d executed-------------\n")
      
print("\n-------------Script executed successfully!-------------\n")
```
